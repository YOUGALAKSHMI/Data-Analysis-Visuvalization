```python
#split dataset in features and target variable
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = df[feature_cols] # Features
y = df.label # Target variable
print(X)
print(y)
```

```python
# Count plot of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title('Distribution of Target Variable')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot of selected features
sns.pairplot(X[feature_cols])
plt.title('Pairplot of Selected Features')
plt.show()
```

```python
# Correlation heatmap of features
plt.figure(figsize=(10, 8))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
# Box plots of features
plt.figure(figsize=(10, 6))
X_boxplot = X.melt()
sns.boxplot(x='variable', y='value', data=X_boxplot)
plt.title('Box Plots of Features')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plots of features against target
for col in feature_cols:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=X[col], y=y)
    plt.title(f'Scatter Plot of {col} vs Outcome')
    plt.xlabel(col)
    plt.ylabel('Outcome')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=X[['glucose', 'bp']].rolling(10).mean())
plt.title('Stacked Line Plot of Glucose and BP vs Age')
plt.xlabel('Age')
plt.ylabel('Values')
plt.legend(['Glucose', 'BP'])
plt.show()
```
