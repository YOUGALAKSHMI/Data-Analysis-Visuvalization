```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,25,0.7)
# plotting the lines
a1 = 4 - 2*x
a2 = 3 - 0.5*x
a3 = 1 -x
# The upper edge of
# polygon
a4 = np.minimum(a1, a2)
# Setting the y-limit
plt.ylim(0, 5)
# Plot the lines
plt.plot(x, a1,
         x, a2,
         x, a3)
# Filling between line a3
# and line a4
plt.fill_between(x, a3, a4, color='red',
                 alpha=0.5)
plt.show()
```

```python
#subplotting
from mpl_toolkits.mplot3d.axes3d import get_test_data
from matplotlib import cm
fig = plt.figure(figsize = (10, 7))
ax = fig.add_subplot(1, 2, 1, projection='3d')
X = np.arange(-25, 15, 0.25)
Y = np.arange(-25,15, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                               linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10)
ax = fig.add_subplot(1, 2, 2, projection='3d')
X, Y, Z = get_test_data(0.05)
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
plt.show()
```

```python
#3D scatter plot
from mpl_toolkits import mplot3d
z = np.random.randint(100, size =(50))
x = np.random.randint(30, size =(50))
y = np.random.randint(50, size =(50))
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z, color = "orange")
plt.title("simple 3D scatter plot")
plt.show()
```

```python
#3D bar plot
fig = plt.figure(figsize=(8, 3))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')
_x = np.arange(3)
_y = np.arange(7)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()
top = x + y
bottom = np.zeros_like(top)
width = depth = 1
ax1.bar3d(x, y, bottom, width, depth, top, shade=True)
ax1.set_title('Shaded')
ax2.bar3d(x, y, bottom, width, depth, top, shade=False)
ax2.set_title('Not Shaded')
plt.show()
```

```python
# Generating random data for demonstration
np.random.seed(42)
data = np.random.randint(1, 100, 20)
```

```python
# Bar Plot (Bar Chart)
categories = ['Category A', 'Category B', 'Category C']
values = data[:3]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

```python
# Line Chart (Line Plot)
x_values = np.arange(1, len(data) + 1)
y_values = data

plt.plot(x_values, y_values)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart Example')
plt.show()
```

```python
# Histogram
plt.hist(data, bins=10, edgecolor='k')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
# Scatterplot
x_values = np.random.rand(20)
y_values = data

plt.scatter(x_values, y_values)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatterplot Example')
plt.show()
```

```python
# Pie Chart
categories = ['Category 0', 'Category 1', 'Category 2', 'Category 3']
values = data[:4]

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```

```python
x = np.linspace(0, 30, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot the two curves
plt.plot(x, y1, label='Sin(x)', color='yellow')
plt.plot(x, y2, label='Cos(x)', color='green')

# Fill the area between the two curves
plt.fill_between(x, y1, y2, where=(y1 < y2), interpolate=True, color='blue', alpha=0.5, label='Between Sin and Cos')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Fill Between Example')

plt.show()
```
