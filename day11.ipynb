```python
img=cv2.imread('/content/3e62f7d2e9dadd4a09691e1ee0500329.jpg')
box,label,c_score=cv.detect_common_objects(img)
output=draw_bbox(img,box,label,c_score)
#output=cv2.cvtColor(output,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(output)
plt.show()
```

```python
from skimage.feature import blob_log
```

```python
fig,ax =plt.subplots(1,2,figsize=(10,5))
ax[0].set_title('Binarized Image',fontsize=15)
ax[0].imshow(img)
blobs =blob_log(img,max_sigma=30,threshold=0.01)
ax[1].imshow(img)
for blob in blobs:
  y,x,area=blob
  ax[1].add_patch(plt.Cricle((x,y),area*np.sqrt(2),color='r',fil=False))
ax[1].set_title('Using LOG',fontsize=15)
plt.tight_layout()
plt.show()
```

```python
total_count=len(blobs)
print("Number of stars counted",total_count)
```
